```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os


img= cv2.imread("note20.png")
if img is None:
    print("Image not loaded")
else:
    print("image loaded")
    #image show 

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))#converts the image from BGR to RGB format because matplotlip expect rgb but it imread loads in bgr
    plt.axis("off")
    plt.show()
    
```

```python
def processes_image(img):
    #enhanced image 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #apply guassian for noise removal 
    blurred = cv2.GaussianBlur(gray, (3, 3), 0)
    return blurred 
    
def preprocess_template(img):
    """Template preprocessing without ximgproc dependency"""
    
    # If image is in color, convert to grayscale
    if len(img.shape) == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        gray = img  # Already grayscale

    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(gray, (3, 3), 0)
    
    return blurred
```

```python
def strict_horizontal_merge(digit_boxes, 
                          max_x_gap=200,  # Max horizontal gap between digits
                          max_y_diff=28,  # Max vertical misalignment
                          height_ratio_thresh=0.5):  # Max height difference (50%)
    if not digit_boxes:
        return []
    
    # Sort by x-coordinate (left to right)
    boxes = sorted(digit_boxes, key=lambda b: b[0])
    
    merged = []
    current = list(boxes[0])
    
    for box in boxes[1:]:
        curr_x1, curr_y1, curr_x2, curr_y2 = current
        box_x1, box_y1, box_x2, box_y2 = box
        
        curr_height = curr_y2 - curr_y1
        box_height = box_y2 - box_y1
        
        # Conditions for Pakistani serial numbers:
        # 1. Reasonable horizontal gap (but can be large for spaced serials)
        # 2. Nearly identical vertical position (strict)
        # 3. Very similar heights
        if (abs(box_x1 - curr_x2) <= max_x_gap and
            abs(box_y1 - curr_y1) <= max_y_diff and
            abs(curr_height - box_height)/max(curr_height, box_height) <= height_ratio_thresh):
            
            # Merge horizontally
            current = [
                min(curr_x1, box_x1),
                min(curr_y1, box_y1),  # Keep original y-position
                max(curr_x2, box_x2),
                max(curr_y2, box_y2)
            ]
        else:
            merged.append(current)
            current = list(box)
    
    merged.append(current)
    
    # Final expansion for prefix letters
    result = []
    for box in merged:
        x1, y1, x2, y2 = box
        result.append((
            max(0, x1 - 70),  # Expand left for prefix
            y1,
            x2,
            y2
        ))
    
    return result
```

```python
def merge_overlapping_boxes(boxes, overlap_threshold=0.1):
    """Merge only boxes that overlap significantly (for same serial number)"""
    if len(boxes) < 2:
        return boxes
    
    # Convert to [x1,y1,x2,y2] and calculate areas
    boxes = np.array(boxes)
    areas = (boxes[:,2] - boxes[:,0] + 1) * (boxes[:,3] - boxes[:,1] + 1)
    
    merged = []
    used = set()
    
    for i in range(len(boxes)):
        if i in used:
            continue
            
        current = boxes[i]
        to_merge = [current]
        
        for j in range(i+1, len(boxes)):
            if j in used:
                continue
                
            # Calculate overlap area
            x1 = max(current[0], boxes[j][0])
            y1 = max(current[1], boxes[j][1])
            x2 = min(current[2], boxes[j][2])
            y2 = min(current[3], boxes[j][3])
            
            overlap_area = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
            min_area = min(areas[i], areas[j])
            
            if overlap_area / min_area > overlap_threshold:
                to_merge.append(boxes[j])
                used.add(j)
        
        # Create merged box
        x1 = min(box[0] for box in to_merge)
        y1 = min(box[1] for box in to_merge)
        x2 = max(box[2] for box in to_merge)
        y2 = max(box[3] for box in to_merge)
        merged.append([x1, y1, x2, y2])
    
    return merged
```

```python
template_path = 'template.png'
output_dir = 'output'
os.makedirs(output_dir, exist_ok=True)

# Load images
note_img = cv2.imread('note20.png')
template_img = cv2.imread('template.png')

# Preprocess
note_processed = processes_image(note_img)
template_processed = preprocess_template(template_img)

# Template matching
res = cv2.matchTemplate(note_processed, template_processed, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = max_loc
h, w = template_processed.shape


# Define adjustments (tweak these values)
blue_offset_x = 10    # move box right by 10 pixels
blue_offset_y = -40   # move box up by 5 pixels
blue_width_adj = -35  # reduce width by 20 pixels
blue_height_adj = -30 # reduce height by 10 pixels

# Calculate new points
adjusted_top_left = (top_left[0] + blue_offset_x, top_left[1] + blue_offset_y)
adjusted_bottom_right = (top_left[0] + w + blue_width_adj, top_left[1] + h + blue_height_adj)

# Draw the adjusted blue box
cv2.rectangle(note_img, adjusted_top_left, adjusted_bottom_right, (255, 0, 0), 2)

# Step 2: Use pixel offset from matched template to find serial number
#  manually measure this offset once based on where the serial appears relative to the phrase)
offset_x = 250 # move right (adjust this manually)
offset_y = -35 # move down  (adjust this manually)

serial_top_left = (top_left[0] + offset_x, top_left[1] + offset_y)
serial_w = 180   # width of serial number block (adjust)
serial_h = 20    # height of serial number block (adjust)

serial_bottom_right = (serial_top_left[0] + serial_w, serial_top_left[1] + serial_h)

# Draw red rectangle on serial number area
cv2.rectangle(note_img, serial_top_left, serial_bottom_right, (0, 0, 255), 2)

# Save results
output_dir = 'Results'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'annotated_note20.png')
cv2.imwrite(output_path, note_img)

# Save coordinates to CSV
csv_path = os.path.join(output_dir, 'coordinates.csv')
with open(csv_path, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['image_name', 'x1', 'y1', 'x2', 'y2'])
    writer.writerow([
        'note20.png',
        serial_top_left[0], serial_top_left[1],
        serial_bottom_right[0], serial_bottom_right[1]
    ])

# Show result
plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(note_img, cv2.COLOR_BGR2RGB))
plt.title("Detected Serial Region")
plt.axis('off')
plt.show()
```
